## **4. DATA PREPROCESSING**

**4.1 Model Performance using "Normalization"** -> It is basically for scaling the data

```python
setup(data=boston,target='medv',normalize=True,normalize_method='zscore')
cm=compare_models()
# normalize_method= { zscore,minmax,maxabs,robust }
# MINMAX --> It scales the data between 0 and 1
# ZSCORE --> It is also known as Standardization.It centers the data around 0
# ROBUST --> Similar to Z Score but uses the median and interquartile(IQR)
```

### **4.2 Model Performance using "Feature Selection"**

Feature selection is basically choosing most important information to make prediction

```python
setup(data=boston,target='medv',feature_selection=True,feature_selection_method='classic')
cm=compare_models()
#feature_selection_threshold=0. 2   ,Don't work with this version of pycaret
#feature_selection_method= {classic,boruta}
# Classic -> Look at every piece of information and decides how important it is before model is trained
# Boruta -> Special Designed for only Random forest model
```

### **4.3 Model Performance using "Outlier Removal"**

Outlier Removal ->It is process in the ML and Data Science in which data points that are different from majority are identified and removed

```python
#Data Visualization can be used to select Outlier Threshold
setup(data=boston,target='medv',remove_outliers=True,outliers_threshold=0.05)
cm=compare_models()
```

### **4.4 Model Performance using "Transformation**

Transformation is applied to ensure the data is in suitable form for the chosen ML algorithm

```python
# It also tell which model type will be good
setup(data=boston,target='medv',transformation=True,transformation_method='yeo-johnson')
cm=compare_models()
```

### **4.5 Model Performance using PCA**
 PCA ->Principal Component Analysis -> It is used to reduce the number of features while retaining the most important features  

```python
setup( data=boston,target='medv',pca=True,pca_method='linear')
cm=compare_models()
```

### **4.6 Model Performance using "Outlier Removal" and "Normalization"**

```python
setup(data=boston,target='medv',remove_outliers=True,outliers_threshold=0.05,normalize=True,normalize_method='zscore')
cm=compare_models()
```

### **4.7 Model Performance using Noramlization + Transformation + Outlier Removal**

```python
setup(data=boston,target='medv',remove_outliers=True,outliers_threshold=0.05,normalize=True,normalize_method='zscore',transformation=True,
          transformation_method='yeo-johnson')
cm=compare_models()
```

## **7.  Plot The Trained Model**

Following Parameters can be plot for a trained model

*   Prediction error plot -'error' -> This plot shows the difference between  predicted value and actual value

*   Learning Curve -'learning' ->Learning curves show how well your model is learning from data and how well it's generalizing to new, unseen data.
*   Validation Curve - 'vc' -> A validation curve helps you find the best setting for a model's hyperparameter by showing how changes in that hyperparameter impact both its performance on known data and its ability to enjoy new, unseen data  .
 Hyperparameters are like Learning rate,No. of trees(In a Random Forest) etc


*  Feature Importance- 'feature'



*   Model Hyerparameter - 'parameter' -> Plot  showing how changing a hyperparameter affects model performance(accuracy)
